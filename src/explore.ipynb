```python
# Your code here
import pandas as pd
import matplotlib.pyplot as plt
import os
from keras.preprocessing import image
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D , Flatten
from keras.callbacks import ModelCheckpoint, EarlyStopping

```

```python

train_dir = "../train"

image_files = os.listdir(train_dir)

cat_images = [file for file in image_files if file.startswith("cat")]

print("Todos los archivos en la carpeta:", image_files)
print("Imágenes de gatos seleccionadas:", cat_images)

cats = []

for i in range(9):
    img_path = os.path.join(train_dir, cat_images[i])  
    img = image.load_img(img_path, target_size=(200, 200))  
    img_array = image.img_to_array(img) 
    img_array /= 255.0 
    cats.append(img_array)

plt.figure(figsize=(12, 8))

for i in range(len(cats)):
    plt.subplot(3, 3, i + 1)
    plt.imshow(cats[i])
    plt.title('Cat')
    plt.axis('off')

# Ajustar el diseño
plt.tight_layout()

# Mostrar las imágenes
plt.show()
```

```python

dog_images=[file for file in image_files if file.startswith("dog")]

cats=[]

for i in range(9):
    img_path = os.path.join("../train", dog_images[i])  
    img = image.load_img(img_path, target_size=(200, 200))  
    img_array = image.img_to_array(img) 
    img_array /= 255.0 
    cats.append(img_array)

plt.figure(figsize=(12, 8))

for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(cats[i])
    plt.title('Cat')
    plt.axis('off')

# Ajustar el diseño
plt.tight_layout()

# Mostrar las imágenes
plt.show()
```

```python
images=[]
y=[]
size=(200,200)

for img_file in image_files:
    img_path = os.path.join("../train", img_file)  
    img = image.load_img(img_path, target_size=size)  
    img_array = image.img_to_array(img) 
    img_array /= 255.0 

    if img_file.startswith("dog"):
        label = 0  
    elif img_file.startswith("cat"):
        label = 1 

    images.append(img_array)
    y.append(label)

images = np.array(images)
y = np.array(y)   

data=(images,y)

```

```python
image_files_test = os.listdir("../test1")
size=(200,200)

test_img=[]
test_y=[]

for img_file in image_files_test:
    img_path = os.path.join("../test1", img_file)  
    img = image.load_img(img_path, target_size=size)  
    img_array = image.img_to_array(img) 
    img_array /= 255.0 

    label="test"
        
    test_img.append(img_array)
    test_y.append(label)

test_img= np.array(test_img)
test_y= np.array(test_y)   

test_data=(test_img,test_y)

```

```python
model = Sequential()
model.add(Conv2D(input_shape = (200,200,3), filters = 64, kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(Conv2D(filters = 64,kernel_size = (3,3),padding = "same", activation = "relu"))
model.add(MaxPool2D(pool_size = (2,2),strides = (2,2)))
model.add(Conv2D(filters = 128, kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(Conv2D(filters = 128, kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(MaxPool2D(pool_size = (2,2),strides = (2,2)))
model.add(Conv2D(filters = 256, kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(Conv2D(filters = 256, kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(Conv2D(filters = 256, kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(MaxPool2D(pool_size = (2,2),strides = (2,2)))
model.add(Conv2D(filters = 512, kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(Conv2D(filters = 512, kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(Conv2D(filters = 512, kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(MaxPool2D(pool_size = (2,2),strides = (2,2)))
model.add(Conv2D(filters = 512, kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(Conv2D(filters = 512, kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(Conv2D(filters = 512, kernel_size = (3,3), padding = "same", activation = "relu"))
model.add(MaxPool2D(pool_size = (2,2),strides = (2,2)))
model.add(Flatten())
model.add(Dense(units = 4096,activation = "relu"))
model.add(Dense(units = 4096,activation = "relu"))
model.add(Dense(units = 2, activation = "softmax"))

```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(images,y, epochs=1)
```
